# Common Signals in Engineering

A signal that appears oten in the models is one whose time rate of change is directly proportional to the signal itself

...

The general form of the signal is:

$$ x(t) = Ce^{at} $$

where $ C $ and $ a $ can be real or complex numbers.

## Case 1: $ C $ and $ a $ are Real
For this case, both $ C $ and $ a $ are real numbers.
### Mathematical Equation

$$ x(t) = Ce^{at} $$

### Explanation
This represents an exponential growth or decay depending on the sign of $ a $.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define time variable
t = np.linspace(-2, 2, 400)

# Case 1: C and a real
C1 = 1
a1 = 1
x1 = C1 * np.exp(a1 * t)

# Create subplots
fig, plt = plt.subplots(figsize=(15, 7))

# Case 1: C and a real
plt.plot(t, x1.real, label='Real part')
plt.plot(t, x1.imag, label='Imaginary part')
plt.set_title('Case 1: C and a Real')
plt.legend();
```

## Case 2: $ C $ is Complex, $ a $ is Imaginary
For this case, $ C $ is a complex number and $ a $ is purely imaginary.
### Mathematical Equation

$$ x(t) = Ce^{j\omega t} $$

where $ C $ is complex, $ j $ is the imaginary unit, and $ \omega $ is a real number representing the frequency.

### Explanation
This represents a sinusoidal signal with an exponential amplitude modulation.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define time variable
t = np.linspace(-2, 2, 400)

# Case 1: C and a real
C1 = 1
a1 = 1
x1 = C1 * np.exp(a1 * t)

# Case 2: C complex, a imaginary
C2 = 1 + 1j
omega = np.pi
x2 = np.abs(C2) * np.exp(1j * omega * t)

# Create subplots
fig, plt = plt.subplots(figsize=(15, 7))

# Case 1: C and a real
plt.plot(t, x2.real, label='Real part')
plt.plot(t, x2.imag, label='Imaginary part')
plt.set_title('Case 2: C Complex, a Imaginary')
plt.legend();
```

## Case 3: Both $ C $ and $ a $ are Complex

For this case, both $ C $ and $ a $ are complex numbers.

### Mathematical Equation

$$ x(t) = Ce^{(a_r + ja_i)t} $$

where $ a_r $ and $ a_i $ are the real and imaginary parts of $ a $ respectively.

### Explanation
This represents a signal that has both exponential growth/decay and oscillatory behavior.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define time variable
t = np.linspace(-2, 2, 400)

# Case 1: C and a real
C1 = 1
a1 = 1
x1 = C1 * np.exp(a1 * t)

# Case 3: Both C and a complex
C3 = 1 + 1j
a3 = 1 + 1j
x3 = C3 * np.exp((a3.real + 1j * a3.imag) * t)

# Create subplots
fig, plt = plt.subplots(figsize=(15, 7))

# Case 1: C and a real
plt.plot(t, x3.real, label='Real part')
plt.plot(t, x3.imag, label='Imaginary part')
plt.set_title('Case 3: Both C and a Complex')
plt.legend();
```

## Explanation
1. **Case 1: $ C $ and $ a $ Real**
   - This case shows exponential growth or decay depending on the sign of $ a $. If $ a $ is positive, the signal grows exponentially. If $ a $ is negative, the signal decays exponentially.
   
2. **Case 2: $ C $ Complex, $ a $ Imaginary**
   - This case represents a sinusoidal signal with an exponential amplitude modulation. The real part and imaginary part oscillate sinusoidally.

3. **Case 3: Both $ C $ and $ a $ Complex**
   - This case represents a signal that has both exponential growth/decay and oscillatory behavior. The real part and imaginary part exhibit combined exponential and sinusoidal characteristics.

By visualizing these cases, you can see how different values of $ C $ and $ a $ affect the signal's behavior.
